```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Load CSV data
data = np.loadtxt('/content/StudentData.csv', delimiter=',', skiprows=1)  # Skip the header row

# Separate features (X) and target variable (y)
X = data[:, 1:-2]  # All columns except 'Serial No.' and 'Chance of Admit'
y = data[:, -1]  # Last column ('Chance of Admit')
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
plt.scatter(X[:, 0], y)  # Assuming the first column in X is the first feature

# Plot the regression line
plt.plot(X[:, 0], model.predict(X), color='red')
plt.xlabel('GRE Score')  # Adjust label based on the feature used
plt.ylabel('Chance of Admit')
plt.title('Linear Regression of Chance of Admit on GRE Score')  # Adjust title accordingly
plt.show()
```
